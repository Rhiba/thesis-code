```python
def print_exp(exp,text,maxlen,num_exps=1):
 
    xs = []
    ys = []
    for i in range(maxlen):
        txt = text[i] if i < len(text) else '<PAD>'
        if not txt == '<PAD>':
            c = text.count(txt)
        else:
            c = maxlen - len(text)
        total = int(exp.count(txt)/c)/num_exps
        xs.append(txt)
        ys.append(total)
        
    objects = xs[::-1]
    y_pos = np.arange(len(objects))
    performance = ys[::-1]

    plt.barh(y_pos, performance, align='center', alpha=0.5)
    plt.yticks(y_pos, objects)
    plt.xlabel('Weight')
    plt.title('Word')

    plt.show()
    
def powerset(iterable):
    "powerset([1,2,3]) --> () (1,) (2,) (3,) (1,2) (1,3) (2,3) (1,2,3)"
    s = list(iterable)
    return chain.from_iterable(combinations(s, r) for r in range(len(s)+1))
    
def get_all_of_cost(cost,word_cost_func,padded_input):
    # first get all combinations of input
    equal_cost = []
    equal_ids = []
    pwset = list(powerset((list(zip(range(len(padded_input)),padded_input)))))
    for pw in pwset:
        words = [p[1] for p in pw]
        ids = [p[0] for p in pw]
        if sum([word_cost_func[w] for w in words]) == cost:
            equal_cost.append(words)
            equal_ids.append(ids)
            
    return equal_cost,equal_ids

#a, b = get_all_of_cost(2,word_cost_func,['an','absurdist','comedy','<PAD>','<PAD>','<PAD>','<PAD>','<PAD>'])
    
```

### explanation calculation (takes about 2 mins)

```python
def func():
    all_exps = []

    embedding, word2index, index2word, index2embedding = load_embedding_function('SST',maxlen,emb_dims,'../../embeddings/')
    input_len = maxlen
    window_size = emb_dims

    shape = (1,maxlen*emb_dims) if ksize == None else (1,ksize,ksize,emb_dims)
    old_i2e = index2embedding.copy()
    new_i2e = (index2embedding+1)/2
    new_i2e[new_i2e==0.5] = 0
    index2embedding = new_i2e

    bayes_model = PosteriorModel(model_name)

    __Embedding = Embedding(word2index, index2word, old_i2e)


    for idx in range(len(ns)):
    #for idx in range(1):
        n = ns[idx]
        #n = 65775
        knn = knns[idx]
        #knn = 10
        X = X_train[n].reshape(1,*X_train[n].shape)
        N = 50
        y_hat = np.argmax(bayes_model.predict(X,n=N))
        y_class = 'Positive.' if y_hat == 1 else 'Negative.'
        y_true_text = 'Positive.' if y_train[n] == 1 else 'Negative.'
        input_shape = X.flatten().shape

        text_list = text_train[n]

        input_ = text_list[:maxlen] + ['<PAD>']*(maxlen-len(text_list))
        padded_input = input_
        X = X_train[n].reshape(*shape)
        X_no_normalise = old_X_train[n].reshape(*shape)

        nearest_neighbors, eq_convex_hull, vertices_convex_hull = [], [], []
        for i in padded_input:
            tmp = __Embedding.nearest_neighbors(i, knn, method='l2')
            nearest_neighbors += [[old_i2e[word2index[w]] for w in tmp[0]]]
            #print(i,nearest_neighbors)
            eq_convex_hull += [[eq.tolist() for eq in ConvexHull(nearest_neighbors[-1]).equations]]
            vertices_convex_hull += [[eq.tolist() for eq in ConvexHull(nearest_neighbors[-1]).vertices]]

        for i in range(len(vertices_convex_hull)):
            vertices_convex_hull[i] = list(itemgetter(*vertices_convex_hull[i])(nearest_neighbors[i]))
        vertices_convex_hull = np.array(vertices_convex_hull)
        minmax_input_bounds = [[np.min(v, axis=0), np.max(v, axis=0)] for v in vertices_convex_hull]

        #print(minmax_input_bounds)

        for nn in range(maxlen):
            if ksize:
                xx = X_no_normalise.reshape(1,maxlen*emb_dims)
                xx = xx[:,nn*window_size:(nn+1)*window_size]
            else:
                xx = np.array(X_no_normalise[:,nn*window_size:(nn+1)*window_size])
            for i, eq in enumerate(eq_convex_hull[nn]):
                w,b = np.array(eq[:-1]).reshape(emb_dims,1), eq[-1]
                dp = np.dot(xx,np.array(w)) + b
                assert dp <= 1e-3, print("The convex hull is NOT consistent! Error at equation {} (input {}), result of xW<=b is zero or negative, {}".format(i, nn, dp))
        print("The convex hull is consistent: each embedded point belongs to one of the respective {} facets equations".format(sum([len(eq) for eq in eq_convex_hull])))

        # since for neurify we need to normalise, fix the bounds for the zero entry PADs to be 0 to (u-l)/2
        new_minmax = []
        for idx, entry in enumerate(minmax_input_bounds):
            # do the neurify normalising
            ls = (entry[0]+1)/2
            us = (entry[1]+1)/2
            word = padded_input[idx]
            if word == '<PAD>': # fix this later to anything not in the embedding
                new_upper = (us-ls)/2
                new_lower = np.array([0.0 for q in range(len(ls))])
            else:
                new_upper = us
                new_lower = ls
            new_entry = [new_lower,new_upper]
            new_minmax.append(new_entry)

        uppers = (np.array([i[1] for i in new_minmax]).flatten())
        lowers = (np.array([i[0] for i in new_minmax]).flatten())
        new_bounds = list(zip(lowers,uppers))
        #print(new_bounds)
        #break

        SAMPLES = 50
        c = 0
        excl_list = ['<PAD>']
        word_cost_func = dict()
        for word in padded_input:
            if not word in word_cost_func.keys():
                # if its a pad or not in the embedding, give it a high cost
                #if word == '<PAD>' or word2index[word] <= 2 or word == 'arliss' or word == 'howard':
                if word in excl_list:
                    word_cost_func[word] = 100
                else:
                    word_cost_func[word] = 1

        #print(word_cost_func)


        explanations = dict()
        explanations['fw+'] = []
        explanations['fw-'] = []
        while c < SAMPLES:
            tmp_weights = bayes_model.sample()
            bayes_model.set_weights(tmp_weights)
            model = bayes_model.model
            tmp_h5 = 'neurify_sst_fc_tmp.h5'
            model.save(tmp_h5)
            subprocess.run(['python3','convert_to_nnet.py',tmp_h5])
            c += 1

            input_shape = X.shape
            prediction = model.predict(X)
            input_ = X.flatten().tolist()
            y_hat = np.argmax(prediction)
            act_y = y_train[n]
            c_hat = np.max(prediction)
            HS_maxlen = 1e7
            verbose = False

            if act_y == y_hat:
                print('Model in fw+')
                fw = 'fw+'
            else:
                print('Model in fw-')
                fw = 'fw-' 
            print("Classifiation for the input is {} (confidence {})".format(y_hat, c_hat))

            tmp_network_nnet = tmp_h5[:-2]+'nnet'

            output_constraints = [y_hat,(1 if y_hat==0 else 0),1e-3]


            h, exec_time, _ = smallest_explanation_knn(bayes_model.model,tmp_network_nnet,padded_input,X,word_cost_func,new_bounds,y_hat,output_constraints,5,HS_maxlen=10000000,verbose=False)
            print("Minimum Size Explanation found {} (size {})".format(h, len(h)/window_size))
            print("Complementary set of Minimum Size Explanation is {}".format([i for i in range(maxlen*emb_dims) if i not in h]))
            print("Execution Time: {}".format(exec_time))

            word_ids = [int(i/emb_dims) for idx,i in enumerate(h) if idx % emb_dims == 0 ]
            sorted(word_ids)
            words_out = [w for idx,w in enumerate(padded_input) if idx in word_ids]
            print(words_out)

            # words out is our first ORE
            # now get all ORES of that cost

            #first get all combinations of that same cost
            cost = sum([word_cost_func[w] for w in words_out])
            print('COST:',cost)
            combs, combs_ids = get_all_of_cost(cost,word_cost_func,padded_input)
            input_bounds = new_bounds
            u_bounds = [float(e[1]) for e in input_bounds]
            l_bounds = [float(e[0]) for e in input_bounds]
            #print(u_bounds)
            #print(l_bounds)
            #print(len(u_bounds))
            #print(len(l_bounds))

            # then check if they are also an ORE (entails)
            ores = []
            for idx,comb in enumerate(combs):
                # convert to indices
                h = []
                for idy, word in enumerate(comb):
                    w_id = combs_ids[idx][idy]
                    start = w_id*emb_dims
                    end = (w_id*emb_dims)+emb_dims
                    for wi in range(start,end):
                        h.append(wi)
                res = entails(h, len(h), input_, len(input_), u_bounds, l_bounds, tmp_network_nnet)
                if res == 0:
                    ores.append(comb)
            if len(words_out) == maxlen:
                ores = [words_out]

            explanations[fw].append(ores)
            print(ores)

        #print(explanations)
        all_exps.append(explanations)
        np.save(f'ORE_bayesian_exps_ALL_UNCERT/exps_idx{n}.npy',explanations)
        del explanations
        del nearest_neighbors
        del tmp_weights
        del model
        del combs
        del combs_ids
        del eq_convex_hull
        del vertices_convex_hull
        del ores

    print(all_exps)
    del all_exps
    del embedding
    del word2index
    del index2word
    del index2embedding
    del bayes_model
    del __Embedding
    del old_i2e
    del new_i2e

func()
```

### prob of word being in any explanation of a single network

```python
ns = [95111,37800,61857,78474,50906,116122,95886,97626,33114,116297,103247,26260,74496,112626,108648]
knns = [1000,1000,1000,1000,1000,100,1000,1000,100,10,6,7,6,50,6]
entropies = dict()
entropies['aeu'] = []
entropies['meu'] = []
entropies['mieu'] = []
entropies['ceu'] = []
pred_uncerts = []
bayes_model = PosteriorModel(model_name)

for idx,n in enumerate(ns):
    explanations = ast.literal_eval(str(np.load(f'ORE_bayesian_exps_ALL_UNCERT/exps_idx{n}.npy',allow_pickle=True)))
    sentence = text_train[n]
    original_class = 'Positive' if y_train[n] == 1 else 'Negative'
    print('Original input:',' '.join(sentence))
    print('Class:',original_class)
    #print('K-nn k-value:',knns[idx])
    print()
    fw = ['fw+','fw-']
    word_count_map = dict()
    word_count_map['fw+'] = dict()
    word_count_map['fw-'] = dict()
    
    for f in fw:
        print('For',f,'models:')
        print()
        exps = explanations[f]
        if len(exps) == 0:
            continue
        for e_set in exps:
            counted_ws = []
            for e in e_set:
                for w in e:
                    if not w in word_count_map[f].keys():
                        word_count_map[f][w] = 0
                    if w not in counted_ws:
                        word_count_map[f][w] += 1
                        counted_ws.append(w)
        #print(word_count_map[f])
        total_nets = len(exps)
        entropy = []
        for w in word_count_map[f].keys():
            q = word_count_map[f][w]/total_nets
            if q == 0 or q == 1:
                h = 0
            else:
                h = (-q*math.log(q,2))-((1-q)*math.log(1-q,2))
            print(w,f'(p={q}), (H={h:.2f})')
            entropy.append(h)
        
        X_tmp = X_train[n].reshape(1,*X_train[n].shape)
        ys_tmp = []
        for ij in range(50):
            bayes_model.model.set_weights(bayes_model.sample())
            ys_tmp.append(bayes_model.model(X_tmp))
        ys_tmp = np.argmax(np.array(ys_tmp),axis=2).flatten()
        print('Prediction variance:',np.var(ys_tmp))    
        print()
        if len(entropy) > 0 and f == 'fw+':
            entropies['aeu'].append(np.mean(entropy))
            entropies['meu'].append(np.max(entropy))
            entropies['mieu'].append(np.min(entropy))
            entropies['ceu'].append(sum(entropy))
            pred_uncerts.append(np.var(ys_tmp))
```

```python
plt.figure()
plt.title('Average Binary Feature Entropy vs Prediction Variance')
plt.scatter(entropies['aeu'],pred_uncerts)
plt.xlabel('Average entropy (uncertainty)')
plt.ylabel('Prediction variance (uncertainty)')

plt.figure()
plt.hist(entropies['aeu'])
plt.show()

plt.figure()
plt.title('Maximum Binary Feature Entropy vs Prediction Variance')
plt.scatter(entropies['meu'],pred_uncerts)
plt.xlabel('Maximum entropy (uncertainty)')
plt.ylabel('Prediction variance (uncertainty)')

plt.figure()
plt.hist(entropies['meu'])
plt.show()

plt.figure()
plt.title('Minimum Binary Feature Entropy vs Prediction Variance')
plt.scatter(entropies['mieu'],pred_uncerts)
plt.xlabel('Minimum entropy (uncertainty)')
plt.ylabel('Prediction variance (uncertainty)')

plt.figure()
plt.hist(entropies['mieu'])
plt.show()

plt.figure()
plt.title('Cumulative Binary Feature Entropy vs Prediction Variance')
plt.scatter(entropies['ceu'],pred_uncerts)
plt.xlabel('Cumulative entropy (uncertainty)')
plt.ylabel('Prediction variance (uncertainty)')

plt.figure()
plt.hist(entropies['ceu'])
plt.show()
```

### Coverage

```python
import collections

for idx,n in enumerate(ns):
    explanations = ast.literal_eval(str(np.load(f'ORE_bayesian_exps_ALL_UNCERT/exps_idx{n}.npy',allow_pickle=True)))
    sentence = text_train[n]
    original_class = 'Positive' if y_train[n] == 1 else 'Negative'
    print('ID',idx)
    print('Original input:',' '.join(sentence))
    print('Class:',original_class)
    print('K-nn k-value:',knns[idx])
    print()
    
    correct_explanations = explanations['fw+']
    for var in ['fw+','fw-']:
        correct_explanations = explanations[var]
        if len(correct_explanations) == 0:
            continue
        coverage_map = dict()
        for exp_list in correct_explanations:
            for exp in exp_list:
                if not str(exp) in coverage_map:
                    coverage_map[str(exp)] = 0
                coverage_map[str(exp)] += 1

        print('For models in',var)
        print("Explanation:",max(coverage_map,key=coverage_map.get),"Covers:",coverage_map[max(coverage_map,key=coverage_map.get)],f"networks ({(coverage_map[max(coverage_map,key=coverage_map.get)]/len(correct_explanations))*100}%)",)
        print()
        print()
```
