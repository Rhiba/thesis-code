```python
import sys
sys.path.append('../shap')
import shap

import os
sys.path.append('../..')
import numpy as np
import deepbayesHF
import deepbayesHF.optimizers as optimizers
from deepbayesHF import PosteriorModel
from deepbayesHF.analyzers import FGSM
from deepbayesHF.analyzers import eps_LRP
from deepbayesHF.analyzers import chernoff_decision_robustness
import tensorflow as tf
from tensorflow.keras.models import *
from tensorflow.keras.layers import *
import cv2
import random
import matplotlib.pyplot as plt
from collections import namedtuple

from joblib import Parallel, delayed
import subprocess
from statistics import mode
import json
import tensorflow as tf

import lime
from lime import lime_image
from skimage.color import gray2rgb
from skimage.segmentation import mark_boundaries
from lime.wrappers.scikit_image import SegmentationAlgorithm
from skimage.color import gray2rgb, rgb2gray, label2rgb # since the code wants color images
```

## Load data and train bayesian network

```python
# Load mnist data (convert to rgb for LIME)
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()

X_train = X_train.reshape(60000,28,28)
X_train = list(map(gray2rgb,X_train))
X_train = np.array(list(map(lambda x: (x/255.).astype("float32"),X_train)))
print(X_train.shape)

X_test = X_test.reshape(10000,28,28)
X_test = list(map(gray2rgb,X_test))
X_test = np.array(list(map(lambda x: (x/255.).astype("float32"),X_test)))
```

```python
opt = optimizers.VariationalOnlineGuassNewton()
likelihood = tf.keras.losses.SparseCategoricalCrossentropy()

model_name = f'MNIST_safety_exp'

inputs = Input(shape=X_train[0].shape)
tmp = Flatten()(inputs)
tmp = Dense(128,activation='relu')(tmp)
tmp = Dense(64,activation='relu')(tmp)
predictions = Dense(10,activation='softmax')(tmp)
model = Model(inputs=inputs,outputs=predictions)

bayes_model = opt.compile(model,loss_fn=likelihood,
                          epochs=25, learning_rate=0.25,
                          inflate_prior=2.0, log_file='tmp/log.txt')

bayes_model.train(X_train,y_train,X_test,y_test)
bayes_model.save(model_name)
```

```python
model_name = f'MNIST_safety_exp'
bayes_model = PosteriorModel(model_name)
y_pred = bayes_model.predict(X_test[:50],n=50)
check_accuracy = tf.keras.metrics.Accuracy(name="train_acc")
check_accuracy(y_test[:50],np.argmax(y_pred,axis=1))
print()
print('Loaded model accuracy:',f'{check_accuracy.result().numpy()*100:.2f}%')
```

## Pick a certain input, and a very uncertain input

```python
N = 50
# pick a correct classified input
n = 0
while True:
    n = np.random.randint(len(y_train))
    n = 56297
    X = X_train[n]
    y_hat = np.argmax(bayes_model.predict(X.reshape(1,*X.shape),n=N))
    y_class = y_hat
    if y_hat == y_train[n]:
        break
        
plt.imshow(X)
print('Prediction:',y_class)
input_shape = X.flatten().shape

print("Index:",n)
X_safe = X
y_safe = y_hat
n_safe = n
```

```python
# pick an uncertain (but correctly classified) input

N = 50
n = 0
while True:
    n = np.random.randint(0,len(y_train))
    print(n)
    n = 43059
    X = X_train[n]
    
    bayes_model.set_weights(bayes_model.sample())
    prediction = [np.array(bayes_model._predict(X.reshape(1,*X.shape)))[0]]
    for i in range(N-1):
        bayes_model.set_weights(bayes_model.sample())
        prediction.append(np.array(bayes_model._predict(X.reshape(1,*X.shape)))[0])
    y_hat_ls = list(np.argmax(prediction,axis=1))
    y_hat = mode(y_hat_ls)
    if y_hat_ls.count(y_hat) < 0.6*50 and y_train[n] == y_hat:
        print(y_hat)
        print(y_hat_ls)
        break

plt.imshow(X)
print('Prediction:',y_hat)
print(y_hat_ls)
input_shape = X.flatten().shape

print("Index:",n)

X_uncert = X
y_uncert = y_hat
n_uncert = n
```

## Verify the safety of each input for the same epsilon

```python
from deepbayesHF.analyzers import chernoff_decision_robustness
# verify that the strongly correctly classified input is safe with an eps=0.02
safe = chernoff_decision_robustness(bayes_model,np.asarray([X_safe]),eps=0.02,output_cls=y_safe,loss=likelihood,verify=False)
print(safe)
```

```python
from deepbayesHF.analyzers import chernoff_decision_robustness
# verify that the uncertain input is unsafe with an eps=0.02

safe_uncert = chernoff_decision_robustness(bayes_model,np.asarray([X_uncert]),eps=0.02,output_cls=y_uncert,loss=likelihood,verify=False)
print(safe_uncert)
```

## Generate the (SHAP) explanations for each input

```python
# get the safe explanation array (n=30) 

background = X_train[np.random.choice(X_train.shape[0],200,replace=False)]

n = n_safe

X_safe_4d = X_safe.reshape(1,*X_safe.shape)

if not os.path.exists(f'exps/exp{n}'):
    os.mkdir(f'exps/exp{n}')
    
input_path = f'exps/exp{n}/X.npy'
background_path = f'exps/exp{n}/bg.npy'
np.save(input_path,X_safe_4d,False)
np.save(background_path,background,False)

iterations = 30
for i in range(iterations):
    subprocess.Popen(['python3','get_SHAP_exp.py',str(i),model_name,input_path,f'exps/exp{n}',background_path])
        
full = False 
while not full:
    if len([name for name in os.listdir(f'exps/exp{n}') if os.path.isfile(os.path.join(f'exps/exp{n}', name))]) == iterations+2:
        full = True
```

```python
# display the explanations from the sampled networks (SAFE)
exps_safe = []
for f in os.listdir(f'exps/exp{n_safe}'):
    if os.path.isfile(os.path.join(f'exps/exp{n_safe}',f)) and not f.startswith('X') and not f.startswith('bg'):
        tmp = np.load(os.path.join(f'exps/exp{n_safe}',f))
        # the shap.image_plot function only works with an outer list, not np array (no idea why, its the same shape)
        tmp = [tmp[i] for i in range(tmp.shape[0])]
        exps_safe.append(tmp)

X_safe = np.load(f'exps/exp{n_safe}/X.npy')
for exs in exps_safe:
    shap.image_plot(exs, -X_safe)
```

```python
# get the UNsafe explanation array (n=30) 

background = X_train[np.random.choice(X_train.shape[0],200,replace=False)]

n = n_uncert

X_uncert_4d = X_uncert.reshape(1,*X_uncert.shape)

if not os.path.exists(f'exps/exp{n}'):
    os.mkdir(f'exps/exp{n}')
    
input_path = f'exps/exp{n}/X.npy'
background_path = f'exps/exp{n}/bg.npy'
np.save(input_path,X_uncert_4d,False)
np.save(background_path,background,False)

iterations = 30
for i in range(iterations):
    subprocess.Popen(['python3','get_SHAP_exp.py',str(i),model_name,input_path,f'exps/exp{n}',background_path])
        
full = False 
while not full:
    if len([name for name in os.listdir(f'exps/exp{n}') if os.path.isfile(os.path.join(f'exps/exp{n}', name))]) == iterations+2:
        full = True
```

```python
# display the explanations from the sampled networks (UNSAFE)
exps_unsafe = []
for f in os.listdir(f'exps/exp{n_uncert}'):
    if os.path.isfile(os.path.join(f'exps/exp{n_uncert}',f)) and not f.startswith('X') and not f.startswith('bg'):
        tmp = np.load(os.path.join(f'exps/exp{n_uncert}',f))
        # the shap.image_plot function only works with an outer list, not np array (no idea why, its the same shape)
        tmp = [tmp[i] for i in range(tmp.shape[0])]
        exps_unsafe.append(tmp)

X_uncert = np.load(f'exps/exp{n_uncert}/X.npy')
for exs in exps_unsafe:
    shap.image_plot(exs, -X_uncert)
```

## SINGLE CLASS analysis

```python
# SAFE INPUT

shape = X_safe.shape
safe_target_class = y_train[n_safe]

# get a distribution of values for each pixel i.e pixel_array[pixel_no] = [list of values over n sampled networks]
safe_pixel_array = []

flat_images = []
for i,exp in enumerate(exps_safe):
    colour_image = exp[safe_target_class].reshape(*shape[1:])
    flat_image = rgb2gray(colour_image).flatten()
    flat_images.append(flat_image)
    
flat_images = np.array(flat_images)

for i in range(28**2):
    safe_pixel_array.append(flat_images[:,i])
    
safe_pixel_array = np.array(safe_pixel_array)
    
safe_means = []
safe_vars = []

for val_list in safe_pixel_array:
    safe_means.append(np.mean(val_list))
    safe_vars.append(np.var(val_list))
    
safe_means = np.array(safe_means)
safe_vars = np.array(safe_vars)

safe_vars_2d = safe_vars.reshape(28,28)
        
X,Y = np.meshgrid(range(28),range(28))

plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, safe_vars_2d, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')

plt.show()
```

```python
# uncert INPUT

shape = X_uncert.shape
uncert_target_class = y_train[n_uncert]

# get a distribution of values for each pixel i.e pixel_array[pixel_no] = [list of values over n sampled networks]
uncert_pixel_array = []

flat_images = []
for i,exp in enumerate(exps_unsafe):
    colour_image = exp[uncert_target_class].reshape(*shape[1:])
    flat_image = rgb2gray(colour_image).flatten()
    flat_images.append(flat_image)
    
flat_images = np.array(flat_images)

for i in range(28**2):
    uncert_pixel_array.append(flat_images[:,i])
    
uncert_pixel_array = np.array(uncert_pixel_array)
    
uncert_means = []
uncert_vars = []

for val_list in uncert_pixel_array:
    uncert_means.append(np.mean(val_list))
    uncert_vars.append(np.var(val_list))
    
uncert_means = np.array(uncert_means)
uncert_vars = np.array(uncert_vars)

uncert_vars_2d = uncert_vars.reshape(28,28)
        
X,Y = np.meshgrid(range(28),range(28))

plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, uncert_vars_2d, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')

plt.show()
```

## ALL CLASS analysis

```python
# SAFE INPUT

shape = X_safe.shape

# get flat array of pixel values per class
flat_per_class_safe = dict()
vars_per_class_safe = dict()
means_per_class_safe = dict()
for c in range(10):
    flat_per_class_safe[c] = []
    vars_per_class_safe[c] = []
    means_per_class_safe[c] = []
    flat_images = []
    for i,exp in enumerate(exps_safe):
        colour_image = exp[c].reshape(*shape[1:])
        flat_image = rgb2gray(colour_image).flatten()
        flat_images.append(flat_image)
        
    flat_images = np.array(flat_images)

    for i in range(28**2):
        flat_per_class_safe[c].append(flat_images[:,i])
        vars_per_class_safe[c].append(np.var(flat_images[:,i]))
        means_per_class_safe[c].append(np.mean(flat_images[:,i]))
    
    flat_per_class_safe[c] = np.array(flat_per_class_safe[c])
    vars_per_class_safe[c] = np.array(vars_per_class_safe[c])
    means_per_class_safe[c] = np.array(means_per_class_safe[c])
```

```python
# UNSAFE INPUT

shape = X_uncert.shape

# get flat array of pixel values per class
flat_per_class_unsafe = dict()
vars_per_class_unsafe = dict()
means_per_class_unsafe = dict()
for c in range(10):
    flat_per_class_unsafe[c] = []
    vars_per_class_unsafe[c] = []
    means_per_class_unsafe[c] = []
    flat_images = []
    for i,exp in enumerate(exps_unsafe):
        colour_image = exp[c].reshape(*shape[1:])
        flat_image = rgb2gray(colour_image).flatten()
        flat_images.append(flat_image)
        
    flat_images = np.array(flat_images)

    for i in range(28**2):
        flat_per_class_unsafe[c].append(flat_images[:,i])
        vars_per_class_unsafe[c].append(np.var(flat_images[:,i]))
        means_per_class_unsafe[c].append(np.mean(flat_images[:,i]))
    
    flat_per_class_unsafe[c] = np.array(flat_per_class_unsafe[c])
    vars_per_class_unsafe[c] = np.array(vars_per_class_unsafe[c])
    means_per_class_unsafe[c] = np.array(means_per_class_unsafe[c])
```

```python
# Compare variance for both inputs over all classes
vmax = 0
for c in range(10):
    vmax = np.max([vmax,np.max(vars_per_class_safe[c]),np.max(vars_per_class_unsafe[c])])
    
for c in range(10):
    print("Explanation for class",c)
    plt.figure()
    ax1 = plt.subplot(1,2,1)
    ax1.imshow(vars_per_class_safe[c].reshape(28,28),cmap='viridis',vmax=vmax)
    ax2 = plt.subplot(1,2,2)
    ax2.imshow(vars_per_class_unsafe[c].reshape(28,28),cmap='viridis',vmax=vmax)
    plt.show()
    print('------------------------')
```

```python
# Look for pixels that have a high mean but low variance on more than one class 
# (ie consistently using the same pixel to predict multiple classes)

# reshape the array such that it is arr[pixel][class] so its easier to work with
flipped_means_per_class_safe = dict()
flipped_vars_per_class_safe = dict()
flipped_means_per_class_unsafe = dict()
flipped_vars_per_class_unsafe = dict()

for c in range(10):
    ms = means_per_class_safe[c]
    for idx,m in enumerate(ms):
        if not idx in flipped_means_per_class_safe:
            flipped_means_per_class_safe[idx] = []
        flipped_means_per_class_safe[idx].append(m)
        
    mus = means_per_class_unsafe[c]
    for idx,mu in enumerate(mus):
        if not idx in flipped_means_per_class_unsafe:
            flipped_means_per_class_unsafe[idx] = []
        flipped_means_per_class_unsafe[idx].append(mu)
        
    vs = vars_per_class_safe[c]
    for idx,v in enumerate(vs):
        if not idx in flipped_vars_per_class_safe:
            flipped_vars_per_class_safe[idx] = []
        flipped_vars_per_class_safe[idx].append(v)
        
    vus = vars_per_class_unsafe[c]
    for idx,vu in enumerate(vus):
        if not idx in flipped_vars_per_class_unsafe:
            flipped_vars_per_class_unsafe[idx] = []
        flipped_vars_per_class_unsafe[idx].append(vu)
        
```

```python
# SAFE first
FRAC = 0.2

safe_max_mean = np.max(np.array([means_per_class_safe[c] for c in range(10)]).flatten())
safe_max_var = np.max(np.array([vars_per_class_safe[c] for c in range(10)]).flatten())
print('Checking for pixels with high mean for more than one class, mean fraction-of-max threshold =',FRAC)
print()
# go through pixel by pixel, and first look for high mean on > 1 class
for px in flipped_means_per_class_safe.keys():
    means = flipped_means_per_class_safe[px]
    c = []
    for m in means:
        if m > safe_max_mean*FRAC:
            c.append(m)
    if len(c) > 1:
        print(f'Pixel coordinate ({int(px/28)},{px%28}):')
        #print(means)
        #print(flipped_vars_per_class_safe[px])
        for idx,var in enumerate(flipped_vars_per_class_safe[px]):
            mp = (means[idx]/safe_max_mean)*100
            mp = "{:7.4f} % of max".format(mp) if mp > 0 else "             n/a"
            print(f"class {idx} variance ({var:.10f}, {(var/safe_max_var)*100:8.4f} % of max) || mean ({means[idx]:11.8f}, {mp})")
            
        print()
```

```python
# UNSAFE
FRAC = 0.2

safe_max_mean = np.max(np.array([means_per_class_safe[c] for c in range(10)]).flatten())
safe_max_var = np.max(np.array([vars_per_class_safe[c] for c in range(10)]).flatten())
print('Checking for pixels with high mean for more than one class, mean fraction-of-max threshold =',FRAC)
print()
# go through pixel by pixel, and first look for high mean on > 1 class
for px in flipped_means_per_class_safe.keys():
    means = flipped_means_per_class_safe[px]
    c = []
    for m in means:
        if m > safe_max_mean*FRAC:
            c.append(m)
    if len(c) > 1:
        print(f'Pixel coordinate ({int(px/28)},{px%28}):')
        #print(means)
        #print(flipped_vars_per_class_safe[px])
        for idx,var in enumerate(flipped_vars_per_class_safe[px]):
            mp = (means[idx]/safe_max_mean)*100
            mp = "{:7.4f} % of max".format(mp) if mp > 0 else "             n/a"
            print(f"class {idx} variance ({var:.10f}, {(var/safe_max_var)*100:8.4f} % of max) || mean ({means[idx]:11.8f}, {mp})")
            
        print()
```
