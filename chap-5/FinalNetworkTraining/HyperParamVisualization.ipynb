```python
import os 


with open('scores.txt') as f:
    content = f.readlines()
lines = [x.strip() for x in content]

mat = []
for line in lines:
    arr = []
    for j in line.split('_'):
        arr.append(j.split('=')[1])
    mat.append(arr)

    
import pandas as pd
import numpy as np
df = pd.DataFrame(np.asarray(mat).astype('float'), columns=['Steps', 'Lr', 'PostVar', 'PriVar', 'Score', 'Steer', 'Right', 'Left'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot('Score', 'Steer', data=df, hue='Right')
plt.show()
```

```python
df = df[df.Score > 0.7]
df = df[df.Score < 0.85]
df = df[df.Steer < 0.10]
df = df[df.Steer > 0.01]
df
```
